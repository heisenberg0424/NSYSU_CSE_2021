```python
import time
import captcha
import pandas as pd
from PIL import Image
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
```

```python
options = Options()
options.add_argument("--headless")
 
chrome = webdriver.Chrome('./chromedriver', options=options)
url='https://selcrs.nsysu.edu.tw/menu1/qrycourse.asp?HIS=1&eng=&in_eng=&IDNO=&ITEM='
chrome.get(url)
chrome.save_screenshot('screenshot.png')
captcha_img = Image.open('screenshot.png')
captcha_img = captcha_img.crop((410,144,530,169))
captcha_img = captcha_img.resize((124,24))
```

```python
department = Select(chrome.find_element_by_name("D1"))
department.select_by_value('B304')
robot = chrome.find_element_by_name('ValidCode')
robot.send_keys(captcha.captcha(captcha_img))
robot.submit()
```

```python
next_page = chrome.window_handles[1]
chrome.switch_to.window(next_page)
#soup = BeautifulSoup(chrome.page_source, 'html.parser')
```

```python
dfs = pd.read_html(chrome.page_source)
df = dfs[0].dropna(axis=1)
df = df.rename(columns=df.loc[2,:])
df = df.drop([0,1,2,df.shape[0]-1,df.shape[0]-2],axis=0)
df.index=[i for i in range(df.shape[0])]
df
```
